task 2: Descriptive Statistics:

```python
#Statistics
print(df.describe())

#mode
print(df['TotalPay'].mode())

#median
print(df['TotalPay'].median())

# minimum value
print(df['TotalPay'].min())

# maximum value
df['TotalPay'].max()

# Calculate range
print(df['TotalPay'].max() - df['TotalPay'].min())

# Calculate standard deviation
print(df['TotalPay'].std())
```

task3 : Data Cleaning...
after looking at the excel sheet (the whole data) i realized the first 36160 values in the BENEFITS column had no value, and the rest acted normally so im not deleting the column, im just going to change the value to 0.

and i'm deleting (droping) the notes and status columns cause they mean nothing to the data and have never been used


```python
df['Benefits'].fillna(0, inplace=True, limit=36160)
df.drop(columns=['Status', 'Notes'], inplace=True)
df.head()
```

task 4 : Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

#histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

#pie chart
plt.figure(figsize=(8, 8))
department_counts = df['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different Departments')
plt.axis('equal')
plt.show()
```

task 5 : Grouped Analysis

```python
#grouping based on JobTitle
grouped_df = df.groupby('JobTitle')['TotalPay'].agg(['mean', 'median', 'std', 'count'])

print(grouped_df)
```

task 6 : Simple Correlation Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute the correlation between TotalPay and BasePay
correlation = df['TotalPay'].corr(df['BasePay'])

# Print the correlation coefficient
print("Correlation coefficient between 'TotalPay' and 'BasePay':", correlation)

#Plot a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPay', y='BasePay', data=df)
plt.title('Scatter Plot: Total Pay vs. BasePay')
plt.xlabel('Total Pay')
plt.ylabel('BasePay')
plt.grid(True)
plt.show()
```
